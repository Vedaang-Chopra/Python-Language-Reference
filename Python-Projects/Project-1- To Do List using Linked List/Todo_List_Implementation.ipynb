```python
def initiateToDoList(file):
    file = open(file,'r')
    tasklist = file.read().splitlines()
    tasklist= [i.strip().split(':') for i in tasklist]
    return tasklist
tasklist = initiateToDoList('inputPS13.txt')
tasklist
```

```python
class Task:
    def __init__(self, task_id, task_desc, status, next) -> None:
        self.task_id=task_id
        self.task_desc=task_desc
        self.status=status
        self.next=next
```

```python
class Linked_List_Auxillary_Functions:
    def __init__(self, head)->None:
        self.head = None
        self.tail = None
        
    
    def set_head(self,first_node):
        self.head=first_node
    
    def set_next_node(self,current_node,new_node):
        current_node.next=new_node
        return new_node
    
    def identify_last_node_in_linked_list(self):
        current_reference_to_tail_of_linked_list=self.head
        while current_reference_to_tail_of_linked_list.next!=None:
            current_reference_to_tail_of_linked_list=current_reference_to_tail_of_linked_list.next
        return current_reference_to_tail_of_linked_list
        
        
    def delete_node(self, curr, prev):
        if curr==self.head:
            self.head=self.head.next
        else:
            prev.next=curr.next
        
        
    def print_all_values_of_linked_list(self):
        current_list_pointer=self.head
        while current_list_pointer is not None:
            print(current_list_pointer.task_id,current_list_pointer.task_desc, end="->")
            current_list_pointer=current_list_pointer.next
        print("None")
```

```python
class Task_List(Linked_List_Auxillary_Functions):
    def __init__(self):
        super().__init__(head=None)
        self.task_id=[]
        self.last_task_id=0
    
    def update_to_complete(self, curr):
        curr.status="Completed"
      
    def update_to_incomplete(self, curr):
        curr.status="Incomplete"
    
    def findWholeWord(self,w):
      return re.compile(r'\b({0})\b'.format(w), flags=re.IGNORECASE).search

    # Appends a Task to the Task List(Linked List)..........
    def addTask(self,task_string:str):                    
        CURRENT_TASK_DESCRIPTION = task_string
        # print("Appending to current Linked List: ", CURRENT_TASK_DESCRIPTION)
        if CURRENT_TASK_DESCRIPTION.strip()=="":
            print("Recieved Empty Task, hence breaking Linked List Insertion")
        else:
            new_node=Task(self.last_task_id+1, CURRENT_TASK_DESCRIPTION, "Incomplete",None)
            self.last_task_id=self.last_task_id+1
            self.task_id.append(self.last_task_id)
            if self.head==None:             
                # print("Empty Linked List, Hence populating Head Pointer Value......")
                super().set_head(new_node)     
                self.tail=self.head
                # print("After Insertion, Current Linked List Pointer Points to: ", self.tail, self.tail.task_desc)
            else:
                if self.tail==None:
                    raise Exception("Error in Insertion logic, Please check. Linked List iterator wasn't assigned Head value")
                else:
                    
                    if self.tail.next==None:
                        self.tail=self.set_next_node(self.tail,new_node)
                        # print("After Insertion, Current Linked List Pointer Points to: ", self.tail, self.tail.task_desc)
                    else:
                        # print("Found Some Error in Linked List Insertion, Hence Identifying the Last node, and inserting new noe after it.")
                        self.tail=super().identify_last_node_in_linked_list()
                        # print("Identified Last node of Linked List is: ", self.tail, self.tail.task_desc)
                        self.tail=super().set_next_node(self.tail,new_node)
                        # print("After Insertion, Current Linked List Pointer Points to: ", self.tail, self.tail.task_desc)
            print("Current Task Appended to Linked List: ", self.tail.task_desc, self.tail.task_id)
            return [self.tail.task_desc, self.tail.task_id]

    def removeTask(self, task_string:str, task_number:int): 
        current_node = self.head
        prev_node=None
        if current_node==None:
            print("Task List Empty!!!!")
        else:
            while current_node!=None:
                # print("Current Data in Node for comparison and current key is: ", current_node.task_desc.strip(), task_string.strip())
                if current_node.task_desc.strip() == task_string.strip() or current_node.task_id == task_number:
                    print('Found Task that has to be deleted.....')
                    self.delete_node(current_node, prev_node)
                    print("Node has been Deleted: ", current_node.task_desc.strip(), current_node.task_id)
                    self.tail=super().identify_last_node_in_linked_list()
                    return current_node.task_id,current_node.task_desc 
                else:
                    prev_node=current_node
                    current_node=current_node.next
            print("Unable to Find the Task, Hence No Deletion Happened.....")
    
    def completeTask(self, task_string:str, task_number:int):
        current_node = self.head
        if current_node==None:
            print("Task List Empty!!!!")
        else:
            while current_node!=None:
                # print("Current Data in Node for comparison and current key is: ", current_node.task_desc.strip(), task_string.strip())
                if current_node.task_desc.strip() == task_string.strip() or current_node.task_id == task_number:
                    print('Found Task that has to be updated to Complete.....')
                    self.update_to_complete(current_node)
                    print("Node has been Updated: ", current_node.task_desc.strip(), current_node.status)
                    # self.tail=super().identify_last_node_in_linked_list()
                    # return current_node.task_id
                    return [current_node.task_id,current_node.task_desc]
                else:
                    current_node=current_node.next
            # print("Unable to Find the Task, Hence No Updation Happened.....")
    
    def incompleteTask(self, task_string:str, task_number:int):
        current_node = self.head
        if current_node==None:
            print("Task List Empty!!!!")
        else:
            while current_node!=None:
                # print("Current Data in Node for comparison and current key is: ", current_node.task_desc.strip(), task_string.strip())
                if current_node.task_desc.strip() == task_string.strip() or current_node.task_id == task_number:
                    print('Found Task that has to be updated to Complete.....')
                    self.update_to_incomplete(current_node)
                    print("Node has been Updated: ", current_node.task_desc.strip(), current_node.status)
                    # self.tail=super().identify_last_node_in_linked_list()
                    # return current_node.task_id
                    return [current_node.task_id,current_node.task_desc ]
                    break
                else:
                    current_node=current_node.next
            # print("Unable to Find the Task, Hence No Updation Happened.....")
        
    def searchTask(self,search_string:str):
        found_task=[]
        current_node = self.head
        if current_node==None:
            print("Task List Empty!!!!")
        else:
          if (search_string).isnumeric():
            while current_node!=None:
                # print("Current Data in Node for comparison and current key is: ", current_node.task_desc.strip(), task_string.strip())
                if current_node.task_id.strip() == search_string:
                    print('Found Task based on task number that was requested.....')
                    print("Current Node data: ", current_node.task_desc.strip(), current_node.status, current_node.task_id)
                    found_task.append([current_node.task_desc.strip(), current_node.status, current_node.task_id])
                    current_node=current_node.next
                else:
                    current_node=current_node.next
          else:
            while current_node!=None:
                # print("Current Data in Node for comparison and current key is: ", current_node.task_desc.lower().strip(), search_string.strip())
                if self.findWholeWord(search_string.strip())(current_node.task_desc.lower().strip()):
                    print('Found Task based on description that has been requested.....')
                    print("Current Node data: ", current_node.task_desc.strip(), current_node.status, current_node.task_id)
                    found_task.append([current_node.task_desc.strip(), current_node.status, current_node.task_id])
                    current_node=current_node.next
                else:
                    current_node=current_node.next
          if len(found_task)==0:
            print("Unable to Find Any Task matching the given Input.....")
            return found_task
          else:
            return found_task
    
    def statusTask(self):
      task_details=[]
      current_list_pointer=self.head
      while current_list_pointer is not None:
            task_details.append([current_list_pointer.task_id,current_list_pointer.task_desc])
            current_list_pointer=current_list_pointer.next
      return task_details
```

```python
def task_list_operations(task_operation, task_descr, Complete_Task_list:Task_List):
    if task_operation=='Add a Task':
        current_task_added=Complete_Task_list.addTask(task_descr)
    elif task_operation=='Remove Task':
        task_id_returned=Complete_Task_list.removeTask(task_descr, -1)
    elif task_operation=='Mark Complete':
        completed_task=Complete_Task_list.completeTask(task_descr, -1)
    elif task_operation=='Mark InComplete':
      incomplete_task=Complete_Task_list.incompleteTask(task_descr, -1)
    elif task_operation=='Task Status':
        task_details=Complete_Task_list.statusTask()
        print("Current Exact Task Information:")
        print(task_details)
    elif task_operation=='Search Task':
        found_task=Complete_Task_list.searchTask(task_descr)
    else:
        raise Exception("Unidentified Task")
```

```python
Complete_Task_list=Task_List()
for i in tasklist:
    print("Current Task Operation:", i[0].strip())
    print("Current Task Details:", i[1].strip())
    task_list_operations(i[0], i[1], Complete_Task_list)
    print()
```
